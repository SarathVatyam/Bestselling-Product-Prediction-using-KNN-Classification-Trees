```python
#Calculate the IQR for the DataFrame
Q1 = Toyota_df.quantile(0.25, numeric_only=True)
Q3 = Toyota_df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

# Identify larger IQRs
threshold = 1.5
larger_IQR_columns = IQR[IQR > threshold * (Q3 - Q1)]

print("Columns with larger IQR:")
print(larger_IQR_columns)

#we can observe that the data is not highly disbursed , thus minimal scope for outliers.
```

```python
#Q3.
# Create dummy variables for the 'Fuel_Type' column
dummies = pd.get_dummies(Toyota_df['Fuel_Type'])

# Concatenate the dummy variables with the original DataFrame
Toyota_df = pd.concat([Toyota_df, dummies], axis=1)

# Print the updated DataFrame
print(Toyota_df[dummies.columns])
```

```python
#Q4.

# randomly sample 50% of the row IDs for training
trainData = Toyota_df.sample(frac=0.5, random_state=1)
# sample 30% of the row IDs into the validation set, drawing only from records
# not already in the training set; 60% of 50% is 30%
validData = Toyota_df.drop(trainData.index).sample(frac=0.6, random_state=1)  
# the remaining 20% rows serve as test
testData = Toyota_df.drop(trainData.index).drop(validData.index)

print('Training   : ', trainData.shape)
print('Validation : ', validData.shape)
print('Test       : ', testData.shape)
print()

# alternative way using scikit-learn
trainData, temp = train_test_split(Toyota_df, test_size=0.5, random_state=1)
validData, testData = train_test_split(temp, test_size=0.4, random_state=1)
print('Training   : ', trainData.shape)
print('Validation : ', validData.shape)
print('Test       : ', testData.shape)
```
